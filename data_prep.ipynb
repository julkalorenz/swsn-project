```python
import pandas as pd
from utils import preprocess_dataframe
from IPython.display import display, HTML
import html
```

```python
def show_text_comparison(original: str, processed: str):
    orig_html = html.escape(original)
    proc_html = html.escape(processed)

    html_content = f"""
    <div style="display:flex; gap:20px; font-family:Arial;">

        <div style="
            flex:1;
            padding:15px;
            border:1px solid #ccc;
            border-radius:6px;
            background:#fafafa;
        ">
            <h3 style="margin-top:0; font-weight:600;">Original</h3>
            <div style="white-space:pre-wrap; font-size:14px; line-height:1.5;">
                {orig_html}
            </div>
        </div>

        <div style="
            flex:1;
            padding:15px;
            border:1px solid #ccc;
            border-radius:6px;
            background:#f5faff;
        ">
            <h3 style="margin-top:0; font-weight:600;">Processed</h3>
            <div style="white-space:pre-wrap; font-size:14px; line-height:1.5;">
                {proc_html}
            </div>
        </div>

    </div>
    """

    display(HTML(html_content))
```

```python
data = pd.read_csv('data/tweets_correction/posts.csv')
processed_df = preprocess_dataframe(data, 'post_text', 'created_at')
row = processed_df.iloc[680]
show_text_comparison(row["post_text"], row["processed"])
```

```python
data = pd.read_csv('data/fake_news/BuzzFeed_fake_news_content.csv')
processed_df = preprocess_dataframe(data, 'text', 'publish_date', author_col='authors')
row = processed_df.iloc[7]
show_text_comparison(row["text"], row["processed"])
```
